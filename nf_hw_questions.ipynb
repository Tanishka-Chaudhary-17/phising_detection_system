```python
# calculate electricity bill using different unit slabs
prev_reading = int(input("Enter previous meter reading: "))
cur_reading = int(input("Enter current meter reading: "))
units = cur_reading - prev_reading
# Slab rates (example)
# 0–100 units → ₹5 per unit
# 101–200 units → ₹7 per unit
# 201–300 units → ₹10 per unit
# Above 300 units → ₹12 per unit
bill_amount = 0
if units <= 100:
    bill_amount = units * 5
elif units <= 200:
    bill_amount = (100 * 5) + (units - 100) * 7
elif units <= 300:
    bill_amount = (100 * 5) + (100 * 7) + (units - 200) * 10
else:
    bill_amount = (100 * 5) + (100 * 7) + (100 * 10) + (units - 300) * 12

print("Units Consumed:", units)
print("Total Bill Amount (INR):", bill_amount)

```
